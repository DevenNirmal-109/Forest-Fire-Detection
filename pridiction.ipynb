```python
import os
import pandas as pd
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
```

```python
model = load_model('ForestFireDetection1.h5')
```

```python
image_folder = r"Dataset\the_wildfire_dataset_2n_version\test\nofire"
```

```python
image_files = [f for f in os.listdir(image_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
```

```python
predictions_list = []
```

```python
for img_name in image_files:
    img_path = os.path.join(image_folder, img_name)
    
    # Step 5: Load and preprocess image
    img = image.load_img(img_path, target_size=(150, 150))  # Adjust to your model's input size
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0  # Normalize if needed

    # Step 6: Predict
    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction)  # or just prediction[0] for regression
    
    predictions_list.append([img_name, predicted_class])
```

```python
# Step 7: Save to CSV
df = pd.DataFrame(predictions_list, columns=["Image", "Predicted Class"])
df.to_csv("predictions_nofirex.csv", index=False)

print("Predictions saved to predictions.csv")
```
